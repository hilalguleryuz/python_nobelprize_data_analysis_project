<span style="font-family: Calibri; font-size: 17px; line-height: 1.3;">
The Nobel Prize is a set of reputable international awards given annually in several categories, including Physics, Chemistry, Medicine, Literature, Peace, and Economic Sciences. Established by the will of Alfred Nobel, the Swedish inventor of dynamite, the prizes were first awarded in 1901. They recognize individuals and organizations that have made significant contributions to humanity in their respective fields. Each Nobel laureate receives a medal, a diploma, and a cash award. The Nobel Prize is widely regarded as one of the highest honors in the world, celebrating outstanding achievements that have had a profound impact on society. </span>

</div>

<ul style="font-family: Calibri; font-size: 16px; line-height: 1.3;">
    <li><b>year</b> - The year the Nobel Prize was awarded</li>
    <li><b>category</b> - The category in which the Nobel Prize was awarded</li>
    <li><b>prize</b> - The name of the nobel prize</li>
    <li><b>motivation</b> - A short description about the subject for which the laureate received the award</li>
    <li><b>prizeShare</b> - Indicates whether the laureate shared the prize with others</li>
    <li><b>laureateID</b> - A unique identifier for the laureate</li>
    <li><b>laureateType</b> - Specifies whether the laureate is an individual or an organization</li>
    <li><b>fullName</b> - The full name of the laureate</li>
    <li><b>birthDate</b> - The birthdate of the laureate</li>
    <li><b>birthCity</b> - The city in which the laureate was born</li>
    <li><b>birthCountry</b> - The country in which the laureate was born</li>
    <li><b>sex</b> - The gender of the laureate</li>
    <li><b>organizationName</b> - The name of the organization that the laureate is related with</li>
    <li><b>organizationCity</b> - The city in which the related organization is located</li>
    <li><b>organizationCountry</b> - The country in which the related organization is located</li>
    <li><b>deathDate</b> - The date of death of the laureate</li>
    <li><b>deathCity</b> - The city in which the laureate died</li>
    <li><b>deathCountry</b> - The country in which the laureate died</li>
</ul>

```python
# Remove relevant columns
df.drop(['prize', 'motivation', 'prize_share', 'laureate_type', 'birth_city', 'organization_name', 'organization_city',
         'death_date', 'death_city', 'death_country'], axis=1, inplace=True)
```

<html>
    <body>
      <strong> <p style="font-size:16px; line-height:1.5;">
***Nobel Ödülü, kişinin doğduğu ülkeye değil, yaşadığı ve çalışmalarını yürüttüğü ülkeye atanır. Bundan dolayı 'birth_country' sütununu değil 'organization_country' sütununu baz alarak bütün soruları yanıtladım. Fakat, 'organization_country' sütununda 911 satırdan 244 satır boş olduğu için (neredeyse 4'te 1'i) boş olan satırları 'birth_country' sütunuyla doldurdum.
   </p></strong>
    </body>
</html>

```python
# Fill missing values in the 'organization_country' column
df['organization_country'].fillna(df['birth_country'], inplace=True)

# Create a new column and extract the country inside parentheses
df['country'] = df['organization_country'].str.extract(r'\((.*?)\)')
```

```python
df.head(10)
```

<html>
    <body>
      <strong> <p style="font-size:16px; line-height:1.5;">
Ayrıca ülke isimlerinde bazı düzenlemeler yaptım. Nobel ödülleri 1901 yılında verilmeye başlandığından dolayı bazı ülkelerin isimleri I.Dünya Savaşı öncesinde belirlenen isimler olarak kalmış. Dolayısıyla aynı ülkenin farklı isimleri bulunuyordu. Ben eski ülke isimlerini güncel isimleriyle değiştirdim.
          </p></strong>
    </body>
</html>



```python
# Fill missing values in the 'country' column
df['country'].fillna(df['organization_country'], inplace=True)

# Replace specific values with their corresponding ones
df['country'] = df['country'].replace({"Federal Republic of Germany":"Germany",
                                                                  "then Germany, now France":"France",
                                                                  "Union of Soviet Socialist Republics":"Russia",
                                                                  "Czechoslovakia":"Czech Republic",
                                                                  "People's Republic of China":"China",
                                                                  "United States of America":"USA",
                                                                  "United Kingdom":"UK"})
```

```python
df['country'].unique()
```

```python
df.head()
```

```python
# Remove the related columns
df.drop(['birth_country','organization_country'], axis=1, inplace=True)
```

```python
# Calculating the total number of Nobel Prizes won by USA per decade
df['usa_winner'] = df['country'] == 'USA'
df['decade'] = (np.floor(df['year'] / 10) * 10).astype(int)
total_usa_winner = df.groupby('decade', as_index=False)['usa_winner'].sum()
total_usa_winner
```

```python
# Set the size of the figure
plt.figure(figsize=(14, 6))

# Create line chart
ax = sns.lineplot(x='decade', y='usa_winner', data=total_usa_winner, marker='o', markersize=10, linewidth=2)

# Start the x-axis from 1895
plt.xlim(1895, None)

# Set axis labels and title
plt.xlabel('Decades', size=11)
plt.ylabel('Number of Nobel Prizes', size=11)
plt.title('Number of USA Nobel Prizes by decades', fontsize=18)

# Change background color
grafik = plt.gca()
grafik.patch.set_facecolor('#f2f2f2')

# Set data labels
for index, row in total_usa_winner.iterrows():
    ax.text(row['decade'], row['usa_winner'] + 2, str(row['usa_winner']), ha="center", fontsize=13)
    
plt.show();

```

```python
# Set age ranges
age_bins = [16, 22, 28, 34, 40, 46, 52, 58, 64, 70, 76, 82, 89]

# Split data by age ranges
df8['age_group'] = pd.cut(df8['age'], bins=age_bins, labels=['16-22', '23-28', '29-34', '35-40', '41-46', '47-52', '53-58', 
                                                             '59-64', '65-70', '71-76', '77-82', '83-89'])

# Set unique categories
categories8 = df8['category'].unique()

# Get all age groups
all_age_groups = ['16-22', '23-28', '29-34', '35-40', '41-46', '47-52', '53-58', 
                  '59-64', '65-70', '71-76', '77-82', '83-89']

# Create a subchart layout with 3 rows and 2 columns
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15), gridspec_kw={'hspace': 0.4, 'wspace': 0.4})

# Give blue color to graphics
dark_blue = sns.color_palette('deep')[0]

# Create a sub-graph for each category 
for i, category in enumerate(categories8):
    ax = axes[i//2, i%2]
    category_df = df8[df8['category'] == category]
    age_counts = category_df['age_group'].value_counts().sort_index()
  
    # Fill in missing age groups
    missing_age_groups = set(all_age_groups) - set(age_counts.index)
    for age_group in missing_age_groups:
        age_counts[age_group] = 0
    
    # Plot the histogram
    sns.histplot(data=category_df, x='age_group', color=dark_blue, ax=ax, bins=len(all_age_groups))
    
    # Expand the x-axis to see all age ranges on the x-axis 
    ax.set_xticks(all_age_groups)
    ax.set_title(f'{category}', size=18)
    ax.tick_params(axis='x', rotation=40)
    ax.set_xlabel(None)
    
# Remove empty graphics
if len(categories8) < len(axes.flat):
    for ax in axes.flat[len(categories8):]:
        ax.set_visible(False)

# Adjust graphs
plt.tight_layout()

plt.show();
```
